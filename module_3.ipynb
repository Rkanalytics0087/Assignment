# Question 5.) Explain the concept of type casting in Python with examples?

```python
# Type casting/type conversion>> The preocess of changing the data type of a value/object. while executing/computation using operators, there can be mismatch between the datatype.
# Implicit Type Casting: Implicit type casting occurs when Python automatically converts one data type to another without any user intervention. This usually happens when performing operations that involve different data types.
# Example
# Implicit type casting
a = 5       # Integer
b = 2.5     # Float

# Python automatically converts 'a' to float before addition
result = a + b
print(result)       # Output: 7.5
print(type(result)) # Output: <class 'float'>

# Explicit Type Casting: Explicit type casting, also known as type conversion, is when the programmer manually converts a variable from one type to another using built-in functions like int(), float(), str(), etc.

# Examples:

# Integer to Float:
a = 10      # Integer
b = float(a)  # Explicitly converting to float
print(b)       # Output: 10.0
print(type(b)) # Output: <class 'float'>

# Float to integer
a = 10.5    # Float
b = int(a)  # Explicitly converting to integer
print(b)       # Output: 10
print(type(b)) # Output: <class 'int'>

# Integer into string
a = "123"   # String
b = int(a)  # Explicitly converting to integer
print(b)       # Output: 123
print(type(b)) # Output: <class 'int'>

# there are limitation
# string like "ritik" cannot convertable into integer or other datat type. if you try this you will get an error of invalid conversion.

invalid_str = "ritik"
try:
    converted_int = int(invalid_str)
except ValueError:
    print("Cannot convert to integer: Invalid string")

# Unintended Conversions: Implicit type casting can sometimes lead to unintended results, especially when dealing with mixed data types.
a = "123"
b = 456
result = a + str(b)
print(result)  # Output: "123456"


```
